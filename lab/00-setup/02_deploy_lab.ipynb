# Deployment Commands

**How to use Web Terminal:**
1. Connect to Serverless
2. View → Cluster Tools → Web Terminal
3. Run commands below

---

## Alternative: UI Deployment

**Deploy Bundle:**
- Repos folder → **⋮** menu → Databricks Asset Bundles → Deploy → Choose `dev`

**Run Job:**
- Workflows → `[dev username] wanderbricks_lab_job` → Run Now

---

## Troubleshooting

**Clean state (if deployment fails):**
- Open Web Terminal (View → Cluster Tools → Web Terminal)
- Navigate to your repo folder
- Run: `rm -rf .databricks/bundle/dev`
- Run: `databricks bundle deploy --target dev`
